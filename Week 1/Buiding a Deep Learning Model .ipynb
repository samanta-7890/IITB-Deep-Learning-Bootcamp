```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# Sequential  Model Api

```python
from tensorflow.keras import Sequential
from tensorflow.keras import layers
```

# Understanding the Sequential Model Conceptually

```python
# Define the sequential model with 3 layers
model = keras.Sequential([
     layers.Dense(2, activation='relu', name="layer1"),
     layers.Dense(3, activation='relu', name="layer2"),
     layers.Dense(4, name="layer3")
 ])
    
# Call model on test input
x = tf.ones((3,3))
y = model(x)
```

Sequential model can be created via add method

```python
model = keras.Sequential()
model.add(layers.Dense(2, activation='relu'))
model.add(layers.Dense(3, activation='relu'))
model.add(layers.Dense(4))
```

# Giving name to layers

You can add name argument to name the different layers

```python
model = keras.Sequential(name='my_sequential')
model.add(layers.Dense(2, activation='relu', name='layer1'))
model.add(layers.Dense(3, activation='relu', name='layer2'))
model.add(layers.Dense(4, name='layer3'))
```

# Adding an input

```python
# Method 1
model = keras.Sequential()
model.add(keras.Input(shape=4,))
model.add(layers.Dense(2, activation='relu'))
```

```python
# Method 2
model = keras.Sequential()
model.add(layers.Dense(2, activation='relu', input_shape=(4,)))
```

```python
model.summary()
```

# Example

```python
# define the model
model = keras.Sequential()
model.add(layers.Dense(10, input_shape=(8,)))
model.add(layers.Dense(1))

model.summary()
```

```python
# Deep MLP with 5 hidden layers
from tensorflow.keras.layers import Dense

# Define the model
model = Sequential()
model.add(Dense(100, input_shape=(8,)))
model.add(Dense(80))
model.add(Dense(30))
model.add(Dense(10))
model.add(Dense(5))
model.add(Dense(1))
          
model.summary()          
```

```python
plot_model(model)
```

# Functional Model API(Advanced)

```python
from tensorflow.keras import Input

# define the layers
x_in = Input(shape=(8,))
dense = layers.Dense(64, activation='relu')
x = layers.Dense(10)(x_in)
```

```python
x_out = layers.Dense(1)(x)
```

```python
# example of model defined with functional API
from tensorflow.keras import Model

# define the layers
x_in = Input(shape=(8,))
x = Dense(10)(x_in)
x_out = Dense(1)(x)

# define the model
model = Model(inputs=x_in, outputs=x_out)

# print summary
model.summary()
```

```python
plot_model(model)
```
