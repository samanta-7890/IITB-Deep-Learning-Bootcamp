# Loading Basic Data Science Libraries

```python
import numpy as np # for matrix operations
import pandas as pd # for loading csv files
import matplotlib.pyplot as plt # for data visualization
```

# Loading Data

```python
data = pd.read_csv('heart_disease.csv')
```

```python
data.head() # head() displays the first 5 rows by default
```

```python
data.shape # This gives the shape of the dataframe(i.e. no of rows, no of columns)
```

# Separating Input & Target Variables

```python
x = data.drop('target', axis=1) # Input Variables, axis=1 indicates a column is dropped
y = data['target'] # Target Variable
```

# Splitting Into Training & Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# find the number of input features
x_train.shape[1]
```

# Model Building

```python
# imports
import tensorflow as tf # Importing tensorflow libraries
from tensorflow import keras # Importing keras from tensorflow

from tensorflow.keras import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
```

# Model Creation / Definition

```python
# Building the model
model = Sequential()
model.add(Dense(32, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

# Model Compilation

```python
# compiling the model
from tensorflow.keras.optimizers import RMSprop
optimizer = RMSprop(0.001) # here we have set our learning_rate as 0.001
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# printing the summary of the model
model.summary()
```

```python
# plotting the model
from tensorflow.keras.utils import plot_model
plot_model(model)
```

# History

```python
history = model.fit(x_train, y_train, validation_split=0.2, batch_size=10, epochs=200, verbose=1,)
```

# Model Evaluation

```python
model.evaluate(x_test, y_test)
```

# Model Accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Train', 'Validation'])
plt.show()
```

# Model Loss

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['Train', 'Validation'])
plt.show()
```
